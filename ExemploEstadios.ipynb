```python
# Estabelece um ponto central do mapa
center_lat = -14.24007
center_lon = -53.18050

# Todos os dados dos estagios na variavel consulta
consulta = list(estadios.find())
```

```python
# Converte em dataframe
df_convertido = pd.DataFrame(consulta)

# Imprime a primeira linha do dataframe
df_convertido.iloc[0]
```

```python
# Cria um mapa a partir do ponto central com 800 x 600
locationmap = folium.Map(location=[center_lat, center_lon], zoom_start=5, width=800, height=600)
```

```python
# Recuperando a quantidade de pontos
qtde = len(df_convertido)

qtde
```

```python
for i in range(qtde):
  lng = float(df_convertido.iloc[i]['longitude'])
  lat = float(df_convertido.iloc[i]['latitude'])
  description= 'Estadio: ' + str(df_convertido.iloc[i]['nome']) + '<br>Capacidade: ' + str(df_convertido.iloc[i]['capacidade'])
  folium.Marker(location=[lat, lng],popup = description, icon=folium.Icon(color='red')).add_to(locationmap)

# Exibindo o mapa com marcações
locationmap
```
